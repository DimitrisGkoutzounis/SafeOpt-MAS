```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create data
x1 = np.random.uniform(-1, 1, 500)
x2 = np.random.uniform(-1, 1, 500)
x3 = np.random.uniform(-1, 1, 500)

y_dim_3 = np.sin(x1**3) + np.cos(x2**2) - np.sin(x3)

# Create the plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Scatter plot
sc = ax.scatter(x1, x2, x3, c=y_dim_3, cmap='viridis')

# Add a color bar to understand the color mapping
cbar = plt.colorbar(sc)
cbar.set_label('Value of y_dim_3')

# Adding labels
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('X3')

# Show the plot
plt.show()
```

```python
kern3_1 = GPy.kern.RBF(3)
kern3_2 = GPy.kern.RBF(3)
kern3_3 = GPy.kern.RBF(3)

c = GPy.kern.Coregionalize(3, output_dim=3)



kern_all = kern3_1 + kern3_2 + kern3_3


param_set = safeopt.linearly_spaced_combinations([(-1,1),(-1,1),(-1,1)], 100)

x1 = np.random.uniform(-1.5,1.5,100)
x2 = np.random.uniform(-1.5,1.5,100)
x3 = np.random.uniform(-1.5,1.5,100)
#reshape to 3 dimensions
x1_1 = np.column_stack((x1,x2,x3))
x2_2 = np.column_stack((x2,x3,x1))
x3_3 = np.column_stack((x3,x1,x2))


y_dim_3 = np.sin(x1_1**3) + np.cos(x2_2**2) - np.sin(x3_3)
print(y_dim_3.shape)



model= GPy.models.GPRegression(x1_1,y_dim_3,kern_all)
model.optimize()

mode2 = GPy.models.GPRegression(x2_2,y_dim_3,kern_all)
mode2.optimize()

mode3 = GPy.models.GPRegression(x3_3,y_dim_3,kern_all)
mode3.optimize()


model.plot(projection='2d',visible_dims=[1],legend=False)
mode2.plot(projection='2d',visible_dims=[1],legend=False)
mode3.plot(projection='2d',visible_dims=[1],legend=False)

```

```python
def noise():
    return np.random.normal(0,0.05)
def draw_sample(x):
    return np.random.choice(x.flatten(),1).reshape(-1,1)
```

```python
x1 = np.random.uniform(-1.5,1.5,50)
x2 = np.random.uniform(-1.5,1.5,50)
x3 = np.random.uniform(-1.5,1.5,50)


x1_1 = np.column_stack((draw_sample(x1),draw_sample(x2),draw_sample(x3)))
x2_2 = np.column_stack((draw_sample(x2),draw_sample(x3),draw_sample(x1)))
x3_3 = np.column_stack((draw_sample(x3),draw_sample(x1),draw_sample(x2)))

observation0 = np.sin(x1_1**3) + np.cos(x2_2**2) - np.sin(x3_3)
print(observation0)
```

```python
#create the models

ker1 = GPy.kern.RBF(3)
ker2 = GPy.kern.RBF(3)
ker3 = GPy.kern.RBF(3)

ker = ker1 + ker2 + ker3

gp1 = GPy.models.GPRegression(x1_1,observation0,ker)
gp2 = GPy.models.GPRegression(x2_2,observation0,ker)
gp3 = GPy.models.GPRegression(x3_3,observation0,ker)


param_set = safeopt.linearly_spaced_combinations([(-1,1),(-1,1),(-1,1)], 100)

opt1 = safeopt.SafeOpt(gp1,param_set,[[-np.inf]],beta=1)
```
